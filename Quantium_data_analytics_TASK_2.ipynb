```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Convert the dictionary to a DataFrame
df = pd.DataFrame(data)

# Summarize the data
summary = df.groupby(['LIFESTAGE', 'PREMIUM_CUSTOMER']).size().unstack().fillna(0)

# Calculate proportions
summary_proportion = summary.div(summary.sum(axis=1), axis=0)

# Plotting
plt.figure(figsize=(12, 8))
summary_proportion.plot(kind='bar', stacked=True, color=['#ff9999','#66b3ff','#99ff99'])
plt.title('Proportion of Customers by Affluence and Life Stage')
plt.xlabel('Life Stage')
plt.ylabel('Proportion of Customers')
plt.legend(title='Affluence')
plt.xticks(rotation=45)
plt.show()
```
